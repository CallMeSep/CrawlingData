```python
import requests 
from urllib.request import Request, urlopen
params = { 'src': 'tree'}
x = requests.get("https://tiki.vn/sach-tieng-anh/c320", params,headers={'User-Agent': 'Mozilla/5.0'})
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(x.text) 
# print(soup.prettify())
```

```python
tag = soup.body 
print(tag.attrs)
```

```python
tensach = soup.find_all("div", {"class": "name"})
gia = soup.find_all('div', {'class' : 'price-discount'})
link = soup.find_all("a", {"class": "style__ProductLink-sc-7xd6qw-2 fHwskZ product-item"})
# giam
```

```python
len(link)
```

```python
urls = []
for x in link:
#     print(x['href'])
    urls.append('https://tiki.vn'+ x['href'])
```

```python
len(urls)
```

```python
books = []
for url in urls:
    params = { 'src': 'tree'}
    x = requests.get(url, params,headers={'User-Agent': 'Mozilla/5.0'})
    print(url)
#     soup = BeautifulSoup(x.text) 
    soup = BeautifulSoup(x.content,"lxml")
    try:
        tensach = soup.find_all('h1', {'class' : 'title'})[0].get_text()
        tacgia = soup.find_all('a', {'data-view-id' : 'pdp_details_view_author'})[0].get_text()
        gia = soup.find_all('div', {'class' : 'product-price__current-price'})[0].get_text()
        books.append([tensach,tacgia,gia])
        print(f'{tensach} {tacgia} {gia}')
    except IndexError:
        tensach = soup.find_all('h1', {'class' : 'title'})[0].get_text()
        tacgia = 'null'
        gia = soup.find_all('div', {'class' : 'product-price__current-price'})[0].get_text()
        books.append([tensach,tacgia,gia])
        print(f'{tensach} {tacgia} {gia}')
    print()
    
```

```python
import pandas as pd
df = pd.DataFrame(books, columns =['Title', 'Tac gia', 'Gia'])
(df)
```
